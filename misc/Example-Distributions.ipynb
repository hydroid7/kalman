```julia
using Plots, Distributions, LaTeXStrings
Plots.scalefontsizes(1.6)
default(dpi = 300)
```

```julia
y1 = Normal(2, 0.2)
range = 0:0.01:3
p = plot(ylim = (0, 10), title = "Beispiel: Positionsbestimmung", xlabel = L"Position", ylabel = L"P(\cdot{})", xticks = 0:0.5:5)
plot!(range, pdf.(y1, range), linestyle = :dash, label = L"y_1")
png("plot-beispiel-messung1.png")
p
```

```julia
y2 = Normal(1.5, 0.1)
plot!(range, pdf.(y2, range), label = L"y_2")
png("plot-beispiel-messung2.png")
p
```

```julia
μ = (0.1 / (0.1 + 0.2)) * 2 + (0.2 / (0.1 + 0.2)) * 1.5
@show μ
K = 0.2/(0.1 + 0.2)
@show K
σ = 0.1 - K * 0.2
@show σ
fusion = Normal(μ, σ^2)

plot!(range, pdf.(fusion, range), label = "Korrektur")
png("plot-beispiel-korrektur.png")
p
```
